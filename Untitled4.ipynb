Decision Tree vs Random Forest

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# create dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# train decision tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

# train random forest
rf = RandomForestClassifier(n_estimators=100, max_depth=5)
rf.fit(X_train, y_train)

# predict
dt_pred = dt.predict(X_test)
rf_pred = rf.predict(X_test)

# accuracy
print("DT Test Accuracy:", accuracy_score(y_test, dt_pred))
print("RF Test Accuracy:", accuracy_score(y_test, rf_pred))

# overfitting check
print("DT Train Accuracy:", dt.score(X_train, y_train))
print("RF Train Accuracy:", rf.score(X_train, y_train))

# conclusion
# dt may overfit, rf is more stable
```

Random Forest Hyperparameter Tuning

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# create dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# define parameters
params = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5]
}

# apply grid search
grid = GridSearchCV(RandomForestClassifier(), params, cv=5)
grid.fit(X_train, y_train)

# best result
print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)

# test accuracy
best_model = grid.best_estimator_
print("Test Accuracy:", best_model.score(X_test, y_test))

# conclusion
# tuning improves performance and avoids overfitting
```

SVM (Linear) + Decision Boundary

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# create 2D dataset
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=42)

# train svm model
model = SVC(kernel='linear')
model.fit(X, y)

# plot data points
plt.scatter(X[:, 0], X[:, 1], c=y)

# get axis
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid
xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)

# decision boundary
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model.decision_function(xy).reshape(XX.shape)

# plot boundary line
ax.contour(XX, YY, Z, levels=[0])

# show plot
plt.show()

# conclusion
# svm creates a straight line to separate two classes
```

SVM with Different C Values

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# create dataset
X, y = make_classification(n_samples=500, n_features=10, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# try different C values
for c in [0.1, 1, 10]:

    # train model
    model = SVC(kernel='linear', C=c)
    model.fit(X_train, y_train)

    # predict
    pred = model.predict(X_test)

    # accuracy
    print("C =", c, "Accuracy:", accuracy_score(y_test, pred))

# conclusion
# small C -> large margin (may underfit)
# large C -> small margin (may overfit)
```

KNN Distance Comparison

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# create dataset
X, y = make_classification(n_samples=500, n_features=5, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# euclidean distance
knn_eu = KNeighborsClassifier(metric='euclidean')
knn_eu.fit(X_train, y_train)
pred_eu = knn_eu.predict(X_test)

# manhattan distance
knn_man = KNeighborsClassifier(metric='manhattan')
knn_man.fit(X_train, y_train)
pred_man = knn_man.predict(X_test)

# accuracy
print("Euclidean Accuracy:", accuracy_score(y_test, pred_eu))
print("Manhattan Accuracy:", accuracy_score(y_test, pred_man))

# conclusion
# different distance metrics can give different performance
```

KNN with Feature Scaling

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# create 3D dataset (3 features)
X, y = make_classification(n_samples=500, n_features=3, n_informative=3, n_redundant=0, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# knn without scaling
knn1 = KNeighborsClassifier()
knn1.fit(X_train, y_train)
pred1 = knn1.predict(X_test)

# apply scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# knn with scaling
knn2 = KNeighborsClassifier()
knn2.fit(X_train_scaled, y_train)
pred2 = knn2.predict(X_test_scaled)

# accuracy
print("Without Scaling:", accuracy_score(y_test, pred1))
print("With Scaling:", accuracy_score(y_test, pred2))

# conclusion
# scaling improves knn performance
```

PCA Explained Variance

```python
# load libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# load dataset
X, y = load_iris(return_X_y=True)

# apply pca
pca = PCA()
pca.fit(X)

# plot variance
plt.plot(pca.explained_variance_ratio_)
plt.xlabel("components")
plt.ylabel("variance")
plt.title("pca variance")

# show plot
plt.show()

# conclusion
# select components where variance becomes stable
```

Evaluation using Confusion Matrix

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# create dataset
X, y = make_classification(n_samples=500, n_features=10, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# predict
pred = model.predict(X_test)

# confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, pred))

# precision
print("Precision:", precision_score(y_test, pred))

# recall
print("Recall:", recall_score(y_test, pred))

# f1 score
print("F1 Score:", f1_score(y_test, pred))

# conclusion
# these metrics evaluate model performance
```

K-Fold vs Train-Test

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier

# create dataset
X, y = make_classification(n_samples=500, n_features=10, random_state=42)

# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# test accuracy
print("Train-Test Accuracy:", model.score(X_test, y_test))

# k-fold cross validation
scores = cross_val_score(RandomForestClassifier(), X, y, cv=5)

# average accuracy
print("K-Fold Accuracy:", scores.mean())

# conclusion
# k-fold gives more reliable performance
```

KNN with K = 3 to 10

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# create dataset
X, y = make_classification(n_samples=500, n_features=5, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# try different k values
for k in range(3, 11):

    # train model
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # predict
    pred = knn.predict(X_test)

    # accuracy
    acc = accuracy_score(y_test, pred)
    print("K =", k, "Accuracy:", acc)

# conclusion
# best k gives highest accuracy
```

Bagging vs Single Model

```python
# load libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

# create dataset
X, y = make_classification(n_samples=500, n_features=10, random_state=42)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# single model (decision tree)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
pred1 = dt.predict(X_test)

# bagging model
bag = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10)
bag.fit(X_train, y_train)
pred2 = bag.predict(X_test)

# accuracy
print("Single Model Accuracy:", accuracy_score(y_test, pred1))
print("Bagging Accuracy:", accuracy_score(y_test, pred2))

# conclusion
# bagging improves performance and reduces overfitting
```
